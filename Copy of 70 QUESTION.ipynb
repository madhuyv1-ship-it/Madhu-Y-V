```python
# --- Question 1: Crop Classification Decision Logic ---
# [span_2](start_span)Logic based on nitrogen levels <100, 100-250, >250[span_2](end_span)
nitrogen = float(input("Enter soil nitrogen (mg/kg): "))
if nitrogen < 100:
    print("Recommendation: Apply HIGH nitrogen fertilizer")
    print("Cost: ₹2000/bag")
elif 100 <= nitrogen <= 250:
    print("Recommendation: Apply MEDIUM nitrogen fertilizer")
    print("Cost: ₹1000/bag")
else:
    print("No fertilizer needed, soil is healthy")
```

```python
# --- Question 2: Irrigation Schedule Using Nested If-Else ---
# [span_3](start_span)Logic for Rice, Wheat, Sugarcane based on moisture and temp[span_3](end_span)
crop = input("Enter Crop (Rice/Wheat/Sugarcane): ").capitalize()
moisture = float(input("Enter Soil Moisture %: "))
temp = input("Enter Temperature (Hot/Moderate/Cool): ").capitalize()

needs_water = False
if crop == "Rice":
    if (temp == "Hot" and moisture < 60) or (temp in ["Moderate", "Cool"] and moisture < 50):
        needs_water = True
elif crop == "Wheat":
    if (temp == "Hot" and moisture < 40) or (temp in ["Moderate", "Cool"] and moisture < 30):
        needs_water = True
elif crop == "Sugarcane":
    if (temp == "Hot" and moisture < 70) or (temp in ["Moderate", "Cool"] and moisture < 60):
        needs_water = True

if needs_water:
    print(f"Decision: IRRIGATE NOW for {crop}")
else:("print no irrigation needed yet")
```

```python
# --- Question 3: Pest Severity Assessment ---
# [span_4](start_span)Logic: Low (<5 pest & <10% dmg), Medium (5-15 pest OR 10-25% dmg), High (>15 pest OR >25% dmg)[span_4](end_span)
pests = int(input("Pests per plant: "))
damage = float(input("Damage percentage: "))

if pests > 15 or damage > 25:
    print("Severity Level: HIGH. Recommendation: Immediate Action.")
elif (5 <= pests <= 15) or (10 <= damage <= 25):
    print("Severity Level: MEDIUM. Recommendation: Monitor daily.")
elif pests < 5 and damage < 10:
    print("Severity Level: LOW. Recommendation: Routine check.")
```

```python
# --- Question 4: Crop Maturity Determination ---
# [span_5](start_span)Rice 120-150, Wheat 140-180, Cotton 160-200[span_5](end_span)
crop_type = input("Enter crop (Rice/Wheat/Cotton): ").capitalize()
days = int(input("Days since sowing: "))

ready = False
if crop_type == "Rice" and 120 <= days <= 150: ready = True
elif crop_type == "Wheat" and 140 <= days <= 180: ready = True
elif crop_type == "Cotton" and 160 <= days <= 200: ready = True

if ready:
    print("Status: READY FOR HARVEST")
else:
    print("Status: Not ready or Overdue")
```

```python
# --- Question 5: Weather-based Farming Decision ---
# [span_6](start_span)Rain > 50, Temp > 38, Wind > 40[span_6](end_span)
rain = float(input("Rainfall (mm): "))
temp = float(input("Temperature (°C): "))
wind = float(input("Wind Speed (km/h): "))

if rain > 50: print("Decision: Delay planting, do field preparation")
elif temp > 38: print("Decision: Delay activities, focus on irrigation")
elif wind > 40: print("Decision: Cancel spray operations")
else: print("Decision: Go ahead with planned activities")
```

```python
# --- Question 6: Soil pH Classification ---
# [span_7](start_span)Ranges: <5.5, 5.5-6.5, 6.5-7.5, 7.5-8.5, >8.5[span_7](end_span)
ph = float(input("Enter Soil pH: "))
if ph < 5.5: print("Highly acidic (needs lime)")
elif 5.5 <= ph < 6.5: print("Acidic (may need amendment)")
elif 6.5 <= ph < 7.5: print("Neutral (ideal)")
elif 7.5 <= ph < 8.5: print("Alkaline (needs sulfur)")
else: print("Highly alkaline (needs major amendment)")
```

```python
# --- Question 7: Yield Prediction ---
# Base 40q/ha. [span_8](start_span)Rain: +2%/10mm > 600, -2%/10mm < 400. Temp: -1%/deg outside 25-30[span_8](end_span)
rain = float(input("Rainfall (mm): "))
temp = float(input("Temperature: "))
yield_val = 40

# Rain adjustment
if rain > 600: yield_val += yield_val * 0.02 * ((rain - 600) / 10)
elif rain < 400: yield_val -= yield_val * 0.02 * ((400 - rain) / 10)

# Temp adjustment
if temp < 25: yield_val -= yield_val * 0.01 * (25 - temp)
elif temp > 30: yield_val -= yield_val * 0.01 * (temp - 30)

print(f"Predicted Yield: {yield_val:.2f} quintals/hectare")
```

```python
# --- Question 8: Fertilizer Selection ---
# [span_9](start_span)Logic based on Low/High N, P, K combinations[span_9](end_span)
n = input("Nitrogen Level (Low/High): ").capitalize()
p = input("Phosphorus Level (Low/High): ").capitalize()
k = input("Potassium Level (Low/High): ").capitalize()

if n == "Low" and p == "Low" and k == "Low": print("Use NPK 10-26-26")
elif n == "Low" and p == "High" and k == "High": print("Use Urea only")
elif n == "High" and p == "Low" and k == "Low": print("Use DAP (18-46-0)")
elif n == "High" and p == "High" and k == "Low": print("Use Muriate of Potash")
else: print("Consult an agronomist for custom mix.")

```

```python
# --- Question 11: Rainfall Data Analysis ---
# [span_12](start_span)List: [45, 52, 120, 156, 180, 165, 140, 135, 98, 65, 30, 28][span_12](end_span)
rainfall = [45, 52, 120, 156, 180, 165, 140, 135, 98, 65, 30, 28]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

total_rain = sum(rainfall)
avg_rain = total_rain / len(rainfall)
max_rain = max(rainfall)
min_rain = min(rainfall)
max_month = months[rainfall.index(max_rain)]
min_month = months[rainfall.index(min_rain)]

print(f"Total: {total_rain}, Avg: {avg_rain:.2f}")
print(f"Max: {max_rain} ({max_month}), Min: {min_rain} ({min_month})")
print("Above average months:", [months[i] for i in range(12) if rainfall[i] > avg_rain])
```

```python
# --- Question 12: Crop Yield Tracking ---
# [span_13](start_span)Yield increases weekly until 80% of max (50q)[span_13](end_span)
max_yield = 50
current_yield_pct = 0
week = 0
# Simulating distinct growth rates for demonstration
growth_rates = [5, 12, 25, 35, 45, 55, 60, 65, 70, 75, 78, 79, 80]

while current_yield_pct < 80:
    # Use list if index exists, else cap at 80 for sim
    pct = growth_rates[week] if week < len(growth_rates) else 80
    amount = (pct / 100) * max_yield
    print(f"Week {week+1}: {amount} q/ha ({pct}%)")
    current_yield_pct = pct
    week += 1
```

```python
# --- Question 29: Cost Benefit ---
# Return dictionary[span_30](end_span)
def analyze_farming_practice(ptype, inv, yield_amt, price, cost):
    revenue = yield_amt * price
    profit = revenue - (inv + cost)
    return {"Practice": ptype, "Revenue": revenue, "Profit": profit}

print(analyze_farming_practice("Organic", 70000, 35, 10000, 5000))
```
